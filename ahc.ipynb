```python
from matplotlib import pyplot as plt
from scipy.cluster.hierarchy import dendrogram, linkage
import numpy as np
import pandas as pd
```

```python
data = pd.read_excel("nilai_tryout.xlsx")
print(data)
```

```python
x = np.asarray(data)
print(x)
```

```python
Z = linkage(x)
```

```python
plt.figure(figsize=(25, 10))
plt.title('Dendrogram Hierarchial Klastering')
plt.xlabel('Data')
plt.ylabel('Jarak')
dendrogram(Z)
plt.show()
```
